```python
# 1.2.1

population_dict = {'California': 38332521,
                   'Texas': 26448193,
                   'New York': 19651127,
                   'Florida': 19552860,
                   'Illinois': 12882135}
population = pd.Series(population_dict)
population
```

```python
area_dict = {'California': 423967, 'Texas': 695662, 'New York': 141297,
             'Florida': 170312, 'Illinois': 149995}
area = pd.Series(area_dict)
area

```

```python
print('population is a Series type: \n', population)
print('area is a Series: \n', area)
```

```python
states = pd.DataFrame({'population': population,
                       'area': area})
print(type(states))
print(states.ndim)
print(states.shape)
```

```python
print(states.iloc[1])
```

```python
print(states['population'])
```

```python
states.index
```

```python
print(states)
```

```python
# slicing and keeping the same dimension
print(states[:1])
print(type(states[:1]))
```

```python
# locating with specific number and reducing one more dimension
print(states.iloc[0])
print(type(states.iloc[0]))
```

```python
# slicing and keeping the same dimension
print(states.iloc[0:1])
print(type(states.iloc[0:1]))
```

```python
#### Several popular ways of creating DataFrame instances
# 1 - from an existing Series, assigning a column name
pd.DataFrame(population, columns=['population'])
# 2 - from existing two Series, using a dictionary-alike
# construction
pd.DataFrame({'population': population,
              'area': area})
# 3 - assigning columns and index while constructing
# an instance
pd.DataFrame(np.random.rand(3, 2),
             columns=['foo', 'bar'],
             index=['a', 'b', 'c'])
```

```python

# 1.2.2
# (1)
print(population)
df1 = pd.DataFrame(population, columns=['population'])
print(df1)
```

```python
# (2)
data = [{'a': i, 'b': 2 * i}
        for i in range(3)]
print(data)
df2 = pd.DataFrame(data)
print(df2)
```

```python
#  NaN: not a number
df3 = pd.DataFrame([{'a': 1, 'b': 2}, {'b': 3, 'c': 4}])
print(df3)
```

```python
#(3)
df4 = pd.DataFrame({'population': population,
              'area': area})
print(df4)
```

```python
# (4)
# popular one
df5 = pd.DataFrame(np.random.rand(3, 2),
             columns=['foo', 'bar'],
             index=['a', 'b', 'c'])
print(df5)

```
