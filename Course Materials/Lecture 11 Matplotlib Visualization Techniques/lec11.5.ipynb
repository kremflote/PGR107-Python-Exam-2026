# 5. Histograms


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.randn(1000)
```

```python
print(data[:20])
```

```python
# 5.1 hist
plt.hist(data)
```

```python
plt.hist(data, bins=30, color='green',alpha= 0.15)
```

```python
plt.hist(data, bins=30, alpha=0.5,
         histtype='stepfilled', color='cyan',
         edgecolor='none') #'stepfilled' generates a lineplot that is by default filled, default is bar
```

```python
# 5.2 an application of hist
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(histtype='stepfilled', alpha=.4, bins=100)#stepfilled

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)
```

```python
kwargs = dict(histtype='stepfilled', alpha=.4, bins=100,color='red')
plt.hist(x1,**kwargs)
plt.hist(x2,**kwargs)
#"treat the key-value pairs in the dictionary as additional named arguments to this function call.
```

```python
x4 = np.random.normal(1,10,1000)
x5 = np.random.normal(-10,10,1000)
plt.hist(x4,color='cyan',alpha = 1)
plt.hist(x5,color='navy',alpha=.65)
```

```python
x6 = np.random.normal(5,10,1000)
x7 = np.random.normal(-5,10,1000)
plt.hist(x6,color='cyan',alpha = .9)
plt.hist(x7,color='navy',alpha=.9)
```
