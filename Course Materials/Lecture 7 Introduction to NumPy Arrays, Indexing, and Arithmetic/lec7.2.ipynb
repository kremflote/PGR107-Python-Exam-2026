## 2.4 Boolean Indexing

```python
names = np.array(['Bob', 'Joe', 'Will', 'Bob', 'Will', 'Joe', 'Joe'])
data = np.random.randn(7, 4)
print(names)
print(data)
```

```python
names == 'Bob'
```

```python
data[names == 'Bob']
```

```python
print(data[names == 'Bob', 2:])
# data[names == 'Bob', 3]
```

```python
names != 'Bob'
data[~(names == 'Bob')]
```

```python
cond = names == 'Bob'
print(cond)
data[~cond]
```

```python
# popular way of giving masks, remember how to use!
mask = (names == 'Bob') | (names == 'Will')
mask
data[mask]
```

```python
data[data < 0] = 0
data
```

```python
# remember how to apply!
data[names != 'Joe'] = 7
data
```

## 2.6 Transposing Arrays and Swapping Axes

```python
# 2.6.1
arr = np.arange(15).reshape((3, 5))
print(arr)
print(arr.T)
```

```python
# 2.6.2
arr = np.random.randn(6, 3)
arr
```

```python
np.dot(arr.T, arr)
```

```python
print(np.matmul(arr.T,arr))
```

```python
#2.6.3
arr = np.arange(16).reshape((2, 2, 4))
print(arr)
```

```python
print(arr.ndim,arr.shape)
```

```python
arr = np.arange(15).reshape(5,3,1)
print(arr)
```

```python
arr1 = arr.transpose((1,0,2))
print(arr1)
```
