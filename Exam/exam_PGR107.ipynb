## Question 1: Supernacci Numbers

The Supernacci numbers `s(n)` are defined as:
- `s(1) = 1`
- `s(n)` is the sum of the last `(n - 1) % 3` Supernacci numbers, or `s(n-1)` if that sum would be empty (i.e. when `(n-1) % 3 == 0`).

**Strategy:** Build the sequence iteratively in a list. For each new term, we use slicing `s[-j:]` and the built-in `sum()` to compute the value.

```python
def question1():
    def supernacci(n):
        s = [1]
        for i in range(2, n + 1):
            j = (i - 1) % 3
            # When (n-1) % 3 == 0, the slice s[-0:] would mean s[0:], which is the entire list, not an empty one.
            if j == 0:
                # Sum would be empty, use previous term instead
                s.append(s[-1])
            else:
                # Sum of the last j elements using slicing
                s.append(sum(s[-j:]))

        return s[n - 1]

    # Output the first 12 Supernacci numbers to verify the pattern
    print("n  | s(n)")
    print("---------")
    for n in range(1, 13):
        print(f"{n:<3}| {supernacci(n)}")
```

- In rollDice() `if dice = guess:` declaration changed to comparison `==`
- In rollDice() Last `if dice = guess` missing the colon :
- None of the functions get called, added function calls

**Architecture changes:**
- Added checkDice function instead of the last `if` chain. It was unnecessarily indented and
  generally bad practice since it is inconvenient to both read and maintain. Added `maxAttempts`
  variable to facilitate this function.
- Removed "global" variables `dice` and `guess`. Since our implementation uses a menu wrapper,
  they aren't technically global, but nonlocal. Regardless the program becomes more robust
  by confining variables within the scope they are required.


```python
import random as rnd

def question2():
    def getDiceGuess():
        guessRaw = ''
        while guessRaw not in ('1', '2', '3', '4'):
            print('Guess the number on a 4-sided dice! Enter a number between 1 and 4:')
            guessRaw = input()
        return int(guessRaw)

    def checkDice(dice, guess, attempts, maxAttempts):
        if dice == guess:
            print('Correct! Number of attempts:', attempts)
            return True
        else:
            print('Wrong! Remaining attempts:', maxAttempts - attempts)
            return False

    def rollDice():
        roll = rnd.randint(1, 4)
        return roll

    dice = rollDice()
    maxAttempts = 4

    for attempts in range(1, maxAttempts + 1): # lower boundary inclusive, upper exclusive
        guess = getDiceGuess()
        if checkDice(dice, guess, attempts, maxAttempts):
            return

    print('Game over! The number was', dice)
```
