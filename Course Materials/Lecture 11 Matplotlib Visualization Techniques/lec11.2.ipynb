# 2. Simple Line Plots


```python
# matplotlib inline
import matplotlib.pyplot as plt

print(plt.style.available)
plt.style.use('seaborn-v0_8-poster')
```

```python
import numpy as np
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));


```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

## 2.1 Adjusting the Plot: Line Colors and Styles


```python
plt.plot(x, np.sin(x - 0), color='red')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0')        # Grayscale between 0 and 1
```

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')
```

```python
# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

```python
%matplotlib inline
plt.plot(x, x + 0, '--r')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, '-r')  # dotted red
```

## 2.2 Adjusting the Plot: Axes Limits

```python
plt.plot(x, np.sin(x))

```

```python
plt.plot(x, np.sin(x))
plt.xlim(0, 10)
plt.ylim(-1.2, 1.2)
#

```

```python
plt.plot(x, np.sin(x))

plt.xlim(0, 11)
plt.ylim(-1, 1.2);
```

## 2.3 Labeling Plots


```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")

```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
# plt.axis('equal')

plt.legend()
```

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
       xlabel='x', ylabel='sin(x)',
       title='A Simple Plot')
```
