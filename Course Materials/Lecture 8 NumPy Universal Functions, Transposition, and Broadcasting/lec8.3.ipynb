```python
# 3.2: understanding the rules
# rule1: If the two arrays differ in their number of dimensions,
# the shape of the one with fewer dimensions
# is padded with ones on its leading (left) side.

M = np.ones((2, 3))
a = np.arange(3)
print(M.ndim, M.shape, a.ndim, a.shape)
print(M)
print(a)
```

```python
print(M+a)



```

```python
# Rule 2: If the shape of the two arrays does not match in any dimension,
# the array with shape equal to 1 in that dimension is stretched to
# match the other shape.
a = np.arange(3).reshape((3, 1))
b = np.ones((3,3))+2
print(a.ndim, a.shape, b.ndim, b.shape)
```

```python
print(a,b)
```

```python
print(a+b)
```

```python
# Rule 3: If in any dimension the sizes disagree
# and neither is equal to 1, an error is raised.
import numpy as np
M = np.ones((3, 2))
print('M is 3 by 2 2-d data',M)

```
