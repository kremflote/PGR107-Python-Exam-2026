# 2. introducing lamda

```python
def apply_to_list(some_list, f):
    return [f(x) for x in some_list]
```

```python
ints = [4, 0, 1, 5, 6]
apply_to_list(ints, lambda x: x * 2+1)
```

# 3. iteration


```python
ints = [4, 0, 1, 5, 6]

new_ints = [2*i+1 for i in ints]
print(ints)
print(new_ints)

```

```python
string_list = ['4','0','1','5','6']
new_string_list = [int(i)*2+1 for i in string_list]
print(new_string_list)
print(string_list[:])
```
