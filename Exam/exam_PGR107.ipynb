## Question 1: Supernacci Numbers

The Supernacci numbers `s(n)` are defined as:
- `s(1) = 1`
- `s(n)` is the sum of the last `(n - 1) % 3` Supernacci numbers, or `s(n-1)` if that sum would be empty (i.e. when `(n-1) % 3 == 0`).

**Strategy:** Build the sequence iteratively in a list. For each new term, we use slicing `s[-j:]` and the built-in `sum()` to compute the value.

```python
from prompt_toolkit.utils import to_int


def supernacci(n):
    """
    Computes and returns the n-th Supernacci number.
    Uses slicing and sum().
    
    Rules:
    - s(1) = 1
    - s(n) is the sum of the last (n - 1) % 3 Supernacci numbers, or s(n-1) if the sum is
      empty.
    """
    s = [1]  # s[0] corresponds to s(1)

    for i in range(2, n + 1):
        j = (i - 1) % 3
        # When (n-1) % 3 == 0, the slice s[-0:] would mean s[0:], which is the entire list, not an   empty one.
        if j == 0:
            # Sum would be empty, use previous term instead
            s.append(s[-1])
        else:
            # Sum of the last j elements using slicing
            s.append(sum(s[-j:]))

    return s[n - 1]


# Output the first 12 Supernacci numbers to verify the pattern
print("n  | s(n)")
print("---------")
for n in range(1, 13):
    print(f"{n:<3}| {supernacci(n)}")
```

## Question 2: Gambling Program



```python
import random as rnd

dice = 0
guess = 0

def getDiceGuess():
    global guess
    guessRaw = input()
    while guessRaw not in ('1', '2', '3', '4'):
        print('Guess the number on a 4- sided dice! Enter a number between 1 and 4:')
        print(guessRaw)
        getDiceGuess()
    guess = int(guessRaw)

def rollDice():
    global dice
    dice = rnd.randint(1, 4) # count of dots on a 4-sided dice
    print("Dice rolled: ", dice)

rollDice()
getDiceGuess()

if dice == guess:
    print('You are good!')
    print(dice)
    print(guess)
else:
    print('Sorry! Try to guess again!')
    getDiceGuess()
    if dice == guess:
        print('Congrats! You got it!')
    else:
        print('Sorry! Your third chance!')
    getDiceGuess()
    if dice == guess:
        print('Congrats! You got it!')
    else:
        print('Sorry! You still can try!')
    getDiceGuess()
    if dice == guess:
        print('Congrats! You got it!')
    else:
        print('Nope. You are really bad at this game.')
        print(guess, dice)
```
