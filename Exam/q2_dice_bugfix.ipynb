## Question 2 — Bug Fix: Dice Guessing Game (17 points)

The original code contains **10 bugs**. Each is identified and fixed below.

| # | Location | Bug | Fix |
|---|----------|-----|-----|
| 1 | `rollDice()` | `random.randint(...)` — module aliased as `rnd` but called as `random` | → `rnd.randint(...)` |
| 2 | `rollDice()` | `randint(1, 5)` — a 4-sided dice only has faces 1–4 | → `randint(1, 4)` |
| 3 | `getDiceGuess()` | `guess = input()` — assigns to wrong variable; `guesss` never updated → infinite loop | → `guesss = input()` |
| 4 | `getDiceGuess()` | `return guesss` indented inside the while loop — returns before loop can validate | → dedent `return` to outside the loop |
| 5 | `getDiceGuess()` | Trailing spaces in `('1 ','2 ','3 ','4 ')` — `input()` never matches | → `('1','2','3','4')` |
| 6 | `getDiceGuess()` | Returns string `guesss`, but `dice` is an int — comparison always False | → `return int(guesss)` |
| 7 | `rollDice()` | `getDiceGuess()` never called — first guess compares against global `guess = 0` | → call `getDiceGuess()` to get first guess |
| 8 | `rollDice()` | `if dice = guess:` (×3) — assignment `=` instead of comparison `==` | → `if dice == guess:` |
| 9 | `rollDice()` | Last `if dice = guess` — missing colon `:` at end of line | → `if dice == guess:` |
| 10 | `rollDice()` | Inconsistent use of `guesss` vs `guess` for storing input | → use `getDiceGuess()` consistently for all 4 guesses |

```python
import random as rnd


def getDiceGuess():
    """
    Prompts the user to guess a number between 1 and 4.
    Keeps asking until a valid input is given.
    Returns the guess as an integer.
    """
    guesss = ''                                  # Bug 5 fix: initialize as string, not 0
    while guesss not in ('1', '2', '3', '4'):    # Bug 5 fix: removed trailing spaces
        print('Guess the number on a 4-sided dice! Enter a number between 1 and 4:')
        guesss = input()                         # Bug 3 fix: assign to guesss, not guess
    return int(guesss)                           # Bug 4 fix: return outside loop | Bug 6 fix: return int


def rollDice():
    """
    Rolls a 4-sided dice and gives the player up to 4 guesses.
    """
    dice = rnd.randint(1, 4)     # Bug 1 fix: rnd not random | Bug 2 fix: 4 not 5

    guess = getDiceGuess()       # Bug 7 fix: call getDiceGuess() for first guess
    if dice == guess:
        print('You are good!')
        return

    print('Sorry! Try to guess again!')
    guess = getDiceGuess()       # Bug 10 fix: consistent use of getDiceGuess()
    if dice == guess:            # Bug 8 fix: == not =
        print('Congrats! You got it!')
        return

    print('Sorry! Your third chance!')
    guess = getDiceGuess()       # Bug 10 fix
    if dice == guess:            # Bug 8 fix: == not =
        print('Congrats! You got it!')
        return

    print('Sorry! You still can try!')
    guess = getDiceGuess()       # Bug 10 fix
    if dice == guess:            # Bug 8+9 fix: == not =, added missing colon
        print('Congrats! You got it!')
    else:
        print('Nope. You are really bad at this game.')


rollDice()
```
