```python
import numpy as np

# 3.1.1
A = np.array([1, 2, 3, 4])
B = np.array([5, 6, 7, 8])

# 3.1.2
print("A + B:", A + B)
print("A - B:", A - B)
print("A * B:", A * B)
print("A / B:", A / B)
```

# P8.2

```python
C = np.array([2, 7, 1, 8, 2, 8, 1, 8, 2, 8])

# 3.2.1
print("sum(C):", np.sum(C))

# 3.2.2
print("prod(C):", np.prod(C))

# 3.2.3
print("cumsum(C):", np.cumsum(C))
```
