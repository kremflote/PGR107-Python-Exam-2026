# Visualizing Errors

## Basic error bar

```python
#matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
```

```python
x = np.linspace(0, 10, 50)
dy = 1
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.r');
```

```python
plt.errorbar(x, y, yerr=dy, fmt='o', color='black',
             ecolor='lightgray', elinewidth=3, capsize=0);
```
