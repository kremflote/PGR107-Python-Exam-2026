## Question 1 — Supernacci Numbers (15 points)

The Supernacci numbers `s(n)` are defined as:
- `s(1) = 1`
- `s(n)` is the sum of the last `(n - 1) % 3` Supernacci numbers, or `s(n-1)` if that sum would be empty (i.e. when `(n-1) % 3 == 0`).

**Strategy:** Build the sequence iteratively in a list. For each new term, use slicing `s[-k:]` and the built-in `sum()` to compute the value — no recursion needed.

```python
def supernacci(n):
    """
    Computes and returns the n-th Supernacci number.
    Uses slicing and sum() — no recursion.
    
    Rules:
      s(1) = 1
      s(n) = sum of last k = (n-1) % 3 terms, or s(n-1) if k == 0
    """
    s = [1]  # s[0] corresponds to s(1)

    for i in range(2, n + 1):
        k = (i - 1) % 3
        if k == 0:
            # Sum would be empty — use previous term instead
            s.append(s[-1])
        else:
            # Sum of the last k elements using slicing
            s.append(sum(s[-k:]))

    return s[n - 1]


# --- Output the first 12 Supernacci numbers to verify the pattern ---
print("n  | s(n)")
print("---------")
for n in range(1, 13):
    print(f"{n:<3}| {supernacci(n)}")
```
