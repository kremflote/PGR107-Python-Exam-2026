```python
x = np.array([1,2,3])
print(x)
print(x.ndim,x.shape)

x_2 = x[:,np.newaxis]
print(x_2)
print('adding one more dimension before, then the dimensionality and shape are: ', x_2.ndim, x_2.shape)
```

```python
import timeit
start = timeit.default_timer()
x = np.arange(10)
# print("x     =", x)
print("x + 5 =", x + 5)
print("x - 5 =", x - 5)
print("x * 2 =", x * 2)
print("x / 2 =", x / 2)
print("x // 2 =", x // 2)
print("x ** 2 =", x ** 2)
print("x mod 2 =", x % 2)
print(timeit.default_timer()-start)
```
