## 1.3 Try out yourself

```python
def print_time(start):
    end=timeit.default_timer()
    print(end-start)
    return (end-start)
```

```python
arr = np.arange(100)

start = timeit.default_timer()
arr * arr
time_spent1 = print_time(start)


start = timeit.default_timer()
arr - arr
time_spent2 = print_time(start)

print('The time spent on calculating multiplication is about {} times compared to adding operation: ', time_spent1/time_spent2)

```
