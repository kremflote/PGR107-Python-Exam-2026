```python
list1 = ['a', 'b', 'c', 'd']
list2 = [1, 2, 3, 4]
zipped = list(zip(list1, list2))
print(type(zipped))
print(zipped)
uzipped = list(zip(*zipped))
print(uzipped)
list1 = list(uzipped[0]);
list2 = list(uzipped[1])
list1, list2 = list(uzipped[0]), list(uzipped[1])
print(list1, list2)
```
